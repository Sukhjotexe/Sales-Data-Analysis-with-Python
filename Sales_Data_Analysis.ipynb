#Exploratory Data Analysis

##Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df, palette='Set2') #to get a chart
for bars in ax.containers:
    ax.bar_label(bars) #to get the labels on the chart
plt.show()  #to actually show the chart
```

```python
#Grouping by gender to see the amount spent
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_by_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y= 'Amount', data = sales_by_gender) #making bar chart with this
plt.show()
```

##The graphs show that females have made more purchased and spent more money than man, by a substantial difference

#AGE

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender') #For Seeing the purchases done based on their age group and further divided by Gender
for bars in ax.containers:
    ax.bar_label(bars) #to add values on top
plt.show()
```

```python
#Now to see the total amount spent based on age group and then genders
sales_total_by_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_total_by_age)
plt.show()
```

##The graphs show that most number of buyers are in the age group 26-35 and majority of them are female

#State

```python
#total number of orders placed by states

orders_by_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)}) #for adjusting the size of the chart to show full names
sns.barplot(data = orders_by_state, x = 'State',y= 'Orders')
plt.show()
```

```python
#total amount spent by states

sales_by_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)}) #for adjusting the size of the chart to show full names
sns.barplot(data = sales_by_state, x = 'State',y= 'Amount')
plt.show()
```

##The graph shows that customers in Uttar Pradesh placed the most number of orders, Followed by Maharashtra and Karnataka. Same goes for amount spent by these states. One difference is that for the number of orders Kerala is in Top 10 but for the total amount spent they are not in the top 10.

#Marital Status

```python
#To see on basis on Marital status the number of customers we have who made purchase
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Now we'll want to see the amount spent based on marital status and further divided by gender
sales_by_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_by_marital_status, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

##The graph shows that married people tend to have more orders/purchases done

#Occupation

```python
#Now we will see the number of customers based on the occupation of the customers

sns.set(rc={'figure.figsize':(20,5)}) #to select the right size for this chart
ax = sns.countplot(data = df, x = 'Occupation') 

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Now we will see based on occupation, how much sales were generated by these customers

sales_by_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_by_occ, x = 'Occupation',y= 'Amount')
plt.show()
```

##It's clear in the graphs that most customers who bought the products work in IT sector, followed by Healthcare and Aviations. Also, overall most sales were generated by the IT sectors workers as well, followed by Govt and Automobile employees.

#Product Category

```python
#Now we will see which product category had most numbers of orders

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#We looked at number of orders for each product category, now we'll look at the total sales generated by each product category

sales_by_prod_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_by_prod_cat, x = 'Product_Category',y= 'Amount')
plt.show()

```

##The graphs show that most number of orders were placed for Clothing category, followed by Food and Electronics and Gadgets. However, overall most sales were generated by products of Food category and it's substantially ahead of other categories.

```python
#We will now see the Top 10 most ordered products

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

##Overall, the data shows that most customers for this are married women between the age of 26-35. Most of them are from Uttar Prades, Maharashtra and Karnataka and their occupation is in IT, Healthcare and Aviation industry. Moreover, they are intersted in buying mainly food, clothes and Electronics Gadgets.
